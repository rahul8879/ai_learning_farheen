```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lets create sales data
data = {
    'product': np.random.choice(['A', 'B', 'C'], size=100),
    # lets take sales from normal distribution
    'sales': np.random.normal(loc=500, scale=50, size=100).astype(int),
    'dummy_sales': np.random.randint(100, 1000, size=100),
    'date': pd.date_range(start='2023-01-01', periods=100, freq='D')

}

df = pd.DataFrame(data)
df.head(2)
```

```python
# lets plot sales distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['sales'], bins=20, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# lets plot dummy sales distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['dummy_sales'], bins=20, kde=True)
plt.title('Dummy Sales Distribution')
plt.xlabel('Dummy Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# avg sales by product
avg_sales = df.groupby('product')['sales'].mean().reset_index()
print("Average Sales by Product:")
print(avg_sales)
```
